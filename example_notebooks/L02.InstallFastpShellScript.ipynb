```python
# @title ### **Step 5: Creating a shell script file to run fastp:**
%%writefile run_fastp.sh
#!/bin/bash

# Exit immediately if a command exits with a non-zero status.
set -e

echo "Starting fastp process..."

# --- Configuration for inputs and outputs for fastp ---
SEQ_DIR='drive/MyDrive/Lab_share/Lab_data/class_466/raw.small.RNA-Seq'
OUT_DIR='drive/MyDrive/Lab_share/Lab_data/class_466/clean.small.RNA-Seq'
# Define samples as a bash array
SAMPLES=("YapPool_S01" "YapPool_S02" "YapPool_S03" "YapPool_S09" "YapPool_S10" "YapPool_S11")

# Create the base output directory if it doesn't exist
mkdir -p ${OUT_DIR}
echo "Ensured output directory exists: ${OUTPUT_BASE_DIR}"

# Loop through each sample and run salmon quant
for SAMPLE in "${SAMPLES[@]}"; do
    echo "Processing sample: ${SAMPLE}"
    ./fastp -w 2 \
    --in1  ${SEQ_DIR}/${SAMPLE}_sub_1.fq.gz \
    --in2  ${SEQ_DIR}/${SAMPLE}_sub_2.fq.gz \
    --out1 ${OUT_DIR}/${SAMPLE}_trim_1.fq.gz \
    --out2 ${OUT_DIR}/${SAMPLE}_trim_2.fq.gz  \
    -h ${OUT_DIR}/${SAMPLE}_trim.html \
    -j ${OUT_DIR}/${SAMPLE}_trim.json \
    --cut_tail --cut_window_size 4 --cut_mean_quality 20 --n_base_limit 5 --trim_tail1 2 --trim_tail2 2  > ${OUT_DIR}/${SAMPLE}_trim.log
    echo "Finished processing ${SAMPLE}"
done
```
